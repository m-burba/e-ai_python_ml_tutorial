```python
# One step training
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import torch.optim as optim

# Step 1: Create identifiable data
X = torch.zeros(100, 10, dtype=torch.float32)
for i in range(100):
    for j in range(10):
        X[i, j] = (i + 1) + (j / 10)
y = torch.arange(1, 101, dtype=torch.float32).reshape(-1, 1)

# Step 2: Dataset and DataLoader
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=4, shuffle=False)

# Step 3: Define a simple linear model
model = nn.Linear(10, 1)

# Step 4: Loss and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.0001)

# Step 5: Take one step of training and show gradients
for batch_X, batch_y in loader:
    # Forward pass
    predictions = model(batch_X)
    loss = loss_fn(predictions, batch_y)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()

    # Print gradients
    print("\nGradients before step:")
    for name, param in model.named_parameters():
        if param.grad is not None:
            print(f"{name} grad:\n{param.grad}")

    # Optimization step
    optimizer.step()

    # Print updated parameters
    print("\nUpdated weights and bias:")
    for name, param in model.named_parameters():
        print(f"{name}:\n{param.data}")

    # Show predictions before and after
    print("\nPredictions after one step:")
    print(predictions.detach().squeeze().numpy())
    print("True y:")
    print(batch_y.squeeze().numpy())

    break  # We only run one batch
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Step 1: Create structured identifiable data
X = torch.zeros(100, 10, dtype=torch.float32)
for i in range(100):
    for j in range(10):
        X[i, j] = (i + 1) + (j / 10)

# Normalize X to [0, 1] to avoid exploding gradients
X = (X - X.min()) / (X.max() - X.min())

# Labels: y = 1 to 100
y = torch.arange(1, 101, dtype=torch.float32).reshape(-1, 1)

# Step 2: Dataset and DataLoader
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=8, shuffle=True)

# Step 3: Define a simple linear model
model = nn.Linear(10, 1)

# Step 4: Loss function and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Step 5: Training loop
loss_history = []
for epoch in range(100):
    epoch_loss = 0.0
    for batch_X, batch_y in loader:
        optimizer.zero_grad()
        predictions = model(batch_X)
        loss = loss_fn(predictions, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
    loss_history.append(epoch_loss)
    if epoch % 10 == 0 or epoch == 99:
        print(f"Epoch {epoch+1}: Loss = {epoch_loss:.4f}")

# Step 6: Plot training loss
plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 7: Evaluate predictions
with torch.no_grad():
    predictions = model(X)
    print("\nFinal predictions vs. actual:")
    for i in range(5):
        pred = predictions[i].item()
        true = y[i].item()
        print(f"y_pred = {pred:.2f}, y_true = {true:.2f}")

# Step 8: Print learned weights
print("\nLearned weights (rounded):")
print(model.weight.data.round(decimals=2))
print("Bias:", model.bias.data.round(decimals=2))
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Step 1: Create structured data
X = torch.zeros(100, 10, dtype=torch.float32)
for i in range(100):
    for j in range(10):
        X[i, j] = (i + 1) + (j / 10)

# Normalize X to [0, 1]
X = (X - X.min()) / (X.max() - X.min())

# True labels
y = torch.arange(1, 101, dtype=torch.float32).reshape(-1, 1)

# Dataset and DataLoader
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=8, shuffle=True)

# Model
model = nn.Linear(10, 1)
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Track loss and predictions
loss_history = []
prediction_history = {}  # epoch -> prediction

# Training loop
for epoch in range(100):
    epoch_loss = 0.0
    for batch_X, batch_y in loader:
        optimizer.zero_grad()
        predictions = model(batch_X)
        loss = loss_fn(predictions, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    loss_history.append(epoch_loss)

    # Save prediction at certain intervals
    if epoch % 10 == 0 or epoch == 99:
        with torch.no_grad():
            prediction_history[epoch] = model(X).detach().squeeze().numpy()

    if epoch % 10 == 0 or epoch == 99:
        print(f"Epoch {epoch+1}: Loss = {epoch_loss:.4f}")

# Plot loss
plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot evolution of the learned function
plt.figure(figsize=(10, 5))
true_y = y.squeeze().numpy()
plt.plot(true_y, label="True y", color="black", linewidth=2)

for epoch, pred in prediction_history.items():
    plt.plot(pred, label=f"Epoch {epoch}")

plt.title("Model Predictions Over Training")
plt.xlabel("Sample Index")
plt.ylabel("Predicted y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

# Step 1: Structured input data (100 x 10)
n=20
X = torch.zeros(100, n, dtype=torch.float32)
for i in range(100):
    for j in range(10):
        X[i, j] = (i + 1) + (j / 10)

# Normalize X to [0, 1]
X = (X - X.min()) / (X.max() - X.min())

# Nonlinear target: sqrt(i+1) * sin(2 * i / 100)
original_y = torch.arange(1, 101, dtype=torch.float32)
sqrt_y = torch.sqrt(original_y)
sine = torch.sin(2 * torch.pi * torch.arange(100, dtype=torch.float32) / 100)
y = (sqrt_y * sine).reshape(-1, 1)

# Dataset and DataLoader
dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=8, shuffle=True)

# Step 2: Define a nonlinear model
m=1
mm=128
if m==1:
   # nonlinear model
   model = nn.Sequential(
      nn.Linear(n, mm),
      nn.SiLU(),
      nn.Linear(mm, 1)
  )
else:
   # linear model
   model = nn.Sequential(
      nn.Linear(n, 1),
      nn.ReLU()
   )

# Loss and optimizer
loss_fn = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Track loss and predictions
loss_history = []
prediction_history = {}

# Training loop
etot = 300
for epoch in range(etot):
    epoch_loss = 0.0
    for batch_X, batch_y in loader:
        optimizer.zero_grad()
        predictions = model(batch_X)
        loss = loss_fn(predictions, batch_y)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    loss_history.append(epoch_loss)

    # Save predictions at intervals
    if epoch % (etot/10) == 0 or epoch == etot - 1:
        with torch.no_grad():
            prediction_history[epoch] = model(X).detach().squeeze().numpy()

    if epoch % (etot/10) == 0 or epoch == etot - 1:
        print(f"Epoch {epoch+1}: Loss = {epoch_loss:.4f}")

# Plot training loss
plt.figure(figsize=(8, 4))
plt.plot(loss_history)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot prediction evolution
plt.figure(figsize=(10, 5))
true_y = y.squeeze().numpy()
plt.plot(true_y, label="True target", color="black", linewidth=2)

for epoch, pred in prediction_history.items():
    plt.plot(pred, label=f"Epoch {epoch}")

plt.title("Model Predictions Over Training (Nonlinear Model)")
plt.xlabel("Sample Index")
plt.ylabel("Predicted y")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
